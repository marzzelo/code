```python
import numpy as np
import matplotlib.pyplot as plt

edades = np.array([5, 8, 17, 30, 35, 40, 45, 49, 51, 71, 78, 85, 92, 94, 99,105])

plt.yticks(np.arange(0, 110, 5))
plt.grid(True)
plt.boxplot(edades)
plt.show()
```

```python
(np.min(edades), np.quantile(edades, 0.25), np.median(edades),
np.quantile(edades, 0.75), np.max(edades))
```

## Datos Anormales / Atípicos / Outliers

```python
edades = np.array([6, 
                   41, 51, 51, 52, 54, 57, 60, 61, 65, 
                   95, 100])

plt.grid(True)
plt.yticks(np.arange(0, 110, 5))
plt.boxplot(edades)
plt.show()
```
