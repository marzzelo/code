<h1 style="color:purple">Correlación de Spearman</h1>

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
goles = [2, 15, 3, 8, 0, 11]
asistencias = [4, 7, 8, 6, 1, 13]
print(stats.spearmanr(goles, asistencias)[0])

sns.set_style("whitegrid")
sns.relplot(x=goles, y=asistencias, kind="scatter", s=200, height=4, aspect=1.5)
# draw a linear regression line
sns.regplot(x=goles, y=asistencias, scatter=True, color="red")
plt.show()
```
